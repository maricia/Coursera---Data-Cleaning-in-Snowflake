## Step 7.2 Handle Missing Data with Deletion

**Syntax:** `<expr> IS [ NOT ] NULL`

**Note:** Use logical operators `AND`, `NOT`, `OR` to combine conditions.

```
%%sql -r dataframe_1
SELECT *
  FROM COURSERA.PUBLIC.CUSTOMERS
 WHERE NAME IS NOT NULL
   AND EMAIL IS NOT NULL
   AND DOB IS NOT NULL
   AND PHONE IS NOT NULL;
```

## Step 7.3 Handle Missing Data with Imputation

**Syntax:** `IFF( <condition>, <expr1>, <expr2> )`

Replace NULL values with a default instead of deleting the row.

```
%%sql -r dataframe_2
SELECT ID,
       NAME,
       IFF(EMAIL IS NULL, 'N/A', EMAIL) AS EMAIL,
       IFF(PHONE IS NULL, 'N/A', PHONE) AS PHONE,
       IFF(DOB IS NULL, 'Unknown', DOB) AS DOB,
       ADDRESS,
       REGION,
       COUNTRY,
       COMPANY,
       LASTTRANSACTION
  FROM COURSERA.PUBLIC.CUSTOMERS;
```
