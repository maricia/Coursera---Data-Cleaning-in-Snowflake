## Step 6.2 Use DATEDIFF() to count the number of days between LASTTRANSACTION and CURRENT_DATE()

**Syntax:** `DATEDIFF( <date_or_time_part>, <date_or_time_expr1>, <date_or_time_expr2> )`

**Note:** `CURRENT_DATE()` returns the current date of the system.

```
%%sql -r dataframe_2
SELECT NAME,
       LASTTRANSACTION,
       TO_DATE(LASTTRANSACTION, 'AUTO') AS LAST_TRANSACTION_DATE,
       DATEDIFF('DAY', TO_DATE(LASTTRANSACTION, 'AUTO'), CURRENT_DATE()) AS DAYS_SINCE_LAST_TRANS
  FROM COURSERA.PUBLIC.CUSTOMERS;
```

## Step 6.3 Find Inactive Customers (DaysSinceLastTransaction > 90)

**Note:** Use a SELECT statement with a WHERE clause to filter inactive customers.

```
%%sql -r dataframe_3
SELECT ID,
       SPLIT_PART(TRIM(NAME, ' 0'), ', ', 1) AS FIRST_NAME,
       SPLIT_PART(TRIM(NAME, ' 0'), ', ', 2) AS LAST_NAME,
       EMAIL,
       TO_DATE(DOB, 'MMMM DD, YYYY') AS DOB,
       TO_DATE(LASTTRANSACTION, 'AUTO') AS LAST_TRANSACTION,
       DATEDIFF('DAY', TO_DATE(LASTTRANSACTION, 'AUTO'), CURRENT_DATE()) AS DAYS_SINCE_LAST_TRANS,
       LTRIM(PHONE, '0+') AS STANDARDIZED_PHONE,
       ADDRESS,
       REGION,
       COUNTRY,
       COMPANY
  FROM COURSERA.PUBLIC.CUSTOMERS
 WHERE DATEDIFF('DAY', TO_DATE(LASTTRANSACTION, 'AUTO'), CURRENT_DATE()) > 90;
```
