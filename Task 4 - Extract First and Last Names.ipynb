## Step 4.2 Use SPLIT_PART() to split a Text column into 2 parts using ',' as a delimiter

**Syntax:** `SPLIT_PART(<string>, <delimiter>, <partNumber>)`

**Note:** If the value is negative, the parts are counted backward from the end of the string.

```
%%sql -r dataframe_2
SELECT NAME,
       TRIM(SPLIT_PART(NAME, ',', 1)) AS LAST_NAME,
       TRIM(SPLIT_PART(NAME, ',', 2)) AS FIRST_NAME
  FROM COURSERA.PUBLIC.CUSTOMERS;
```
