```
%%sql -r dataframe_8
SELECT
    ORDERID,
    TRIM(ORDERCITY) AS ORDER_CITY,
    TO_DATE(ORDERDATE, 'MON DD, YYYY') AS ORDERDATE,
    QUANTITY,
    SPLIT_PART(PRODUCT, ',', 2) AS PRODUCT_DESCRIPTION
FROM
    COURSERA.PUBLIC.ORDERS;
```

/* Step 6: Modify the previous query then add :
	Filter dates Between '2023-01-01' and '2023-01-31'
	GROUP BY PRODUCT_DESCRIPTION, ORDER_CITY
	COUNT_ORDERS = COUNT(ORDERID)
	TOTAL_QUANTITY_SOLD = SUM(QUANTITY)
	ORDER BY TOTAL_QUANTITY_SOLD DESC
	LIMIT result to 5
*/

```
%%sql -r dataframe_11
SELECT
    SPLIT_PART(PRODUCT, ',', 2) AS PRODUCT_DESCRIPTION,
    TRIM(ORDERCITY) AS ORDER_CITY,
    COUNT(ORDERID) AS COUNT_ORDERS,
    SUM(QUANTITY) AS TOTAL_QUANTITY_SOLD
FROM
    COURSERA.PUBLIC.ORDERS
WHERE
    TO_DATE(ORDERDATE, 'MON DD, YYYY') BETWEEN '2023-01-01' AND '2023-01-31'
GROUP BY
    PRODUCT_DESCRIPTION, ORDER_CITY
ORDER BY
    TOTAL_QUANTITY_SOLD DESC
LIMIT 5;
```
