## Step 3.2 Use CONCAT() to Visualize Spaces in the "NAME" Column

The function `CONCAT()` is used to facilitate extra spaces detection.

**Syntax:** `CONCAT( <expr1> [ , <exprN> ... ] )`

```
%%sql -r dataframe_2
SELECT NAME,
       CONCAT('|', NAME, '|') AS NAME_WITH_BORDERS
  FROM COURSERA.PUBLIC.CUSTOMERS;
```

## Step 3.3 Use TRIM() to Remove Leading and Trailing Characters from Column "NAME"

**Syntax:** `TRIM( <expr> [, <characters> ] )`

```
%%sql -r dataframe_3
SELECT NAME,
       TRIM(NAME) AS TRIMMED_NAME,
       CONCAT('|', TRIM(NAME), '|') AS TRIMMED_WITH_BORDERS
  FROM COURSERA.PUBLIC.CUSTOMERS;
```
