## Step 5: Extract and clean columns

- **PRODUCT_DESCRIPTION**: Extract product description without color using `SPLIT_PART`
- **ORDER_CITY**: Use `TRIM` to remove leading spaces from `ORDERCITY`
- **ORDER_DATE**: Convert `ORDERDATE` into a Date column using `TO_DATE`
- Keep `ORDERID` and `QUANTITY` as-is

```
%%sql -r dataframe_2
SELECT ORDERID,
       TRIM(ORDERCITY) AS ORDER_CITY,
       TO_DATE(ORDERDATE, 'MON DD, YYYY') AS ORDER_DATE,
       QUANTITY,
       SPLIT_PART(PRODUCT, ',', 1) AS PRODUCT_DESCRIPTION
  FROM COURSERA.PUBLIC.ORDERS;
```

## Step 6: Top 5 sold products per city in Jan 2023

- Filter dates between `2023-01-01` and `2023-01-31`
- Group by `PRODUCT_DESCRIPTION`, `ORDER_CITY`
- `COUNT_ORDERS = COUNT(ORDERID)`
- `TOTAL_QUANTITY_SOLD = SUM(QUANTITY)`
- Order by `TOTAL_QUANTITY_SOLD DESC`
- Limit to 5

```
%%sql -r dataframe_3
SELECT SPLIT_PART(PRODUCT, ',', 1) AS PRODUCT_DESCRIPTION,
       TRIM(ORDERCITY) AS ORDER_CITY,
       COUNT(ORDERID) AS COUNT_ORDERS,
       SUM(QUANTITY) AS TOTAL_QUANTITY_SOLD
  FROM COURSERA.PUBLIC.ORDERS
 WHERE TO_DATE(ORDERDATE, 'MON DD, YYYY') BETWEEN '2023-01-01' AND '2023-01-31'
 GROUP BY PRODUCT_DESCRIPTION, ORDER_CITY
 ORDER BY TOTAL_QUANTITY_SOLD DESC
 LIMIT 5;
```
