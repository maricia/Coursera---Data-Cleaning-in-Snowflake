## Step 8.2 Find Duplicated Customers based on email

**Syntax:** `SELECT ... COUNT(<expr1>) FROM ... GROUP BY ... HAVING <predicate>`

**Note:** Add a `HAVING` clause to filter rows produced by `GROUP BY`.

```
%%sql -r duplicates
SELECT EMAIL,
       COUNT(*) AS DUPLICATE_COUNT
  FROM COURSERA.PUBLIC.CUSTOMERS
 GROUP BY EMAIL
HAVING COUNT(*) > 1;
```

## Step 8.3 Eliminate Duplicated Customers and keep the record with the latest Transaction Date

**Syntax:** `RANK() OVER ( [ PARTITION BY <expr1> ] ORDER BY <expr2> [ ASC | DESC ] )`

**Note:** Use `QUALIFY` instead of a subquery with `WHERE` to filter window function results.

```
%%sql -r dataframe_2
SELECT *
  FROM COURSERA.PUBLIC.CUSTOMERS
QUALIFY RANK() OVER (
            PARTITION BY EMAIL
            ORDER BY TO_DATE(LASTTRANSACTION, 'AUTO') DESC
        ) = 1;
```
