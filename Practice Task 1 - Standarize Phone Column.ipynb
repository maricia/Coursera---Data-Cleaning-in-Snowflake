## Step 1.2 Use LTRIM to remove `+` and `0` from the PHONE column

**Syntax:** `LTRIM( <expr> [, <characters> ] )`

```
%%sql -r dataframe_1
SELECT PHONE,
       LTRIM(PHONE, '+0') AS STANDARDIZED_PHONE
  FROM COURSERA.PUBLIC.CUSTOMERS;
```
